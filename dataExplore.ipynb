```python
#import pdb
for day in null_groups.groups.keys():
    columns = dataset.columns.values.tolist()
    not_null_columns = dataset.groupby(dataset.Time.dt.date).get_group(str(day)).dropna(axis=1).columns.values.tolist()
    columns_to_be_filled = set(columns) - set(not_null_columns)
    columns_to_be_filled = map(lambda x: x, columns_to_be_filled)
   
    means = dataset.groupby(dataset.Time.dt.date).get_group(str(day)).mean()
    means = means[columns_to_be_filled]
    idxs = null_groups.get_group(str(day)).index.values.tolist()
    #pdb.set_trace()
    dataset.loc[idxs, columns_to_be_filled] = means.tolist()
```
